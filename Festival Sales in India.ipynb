```python
/// Grouping by Genders 
```

```python
ax=sns.countplot(x='Gender',data=df)
#///Adding values to the graphs 
for i in ax.containers:
    ax.bar_label(i)
```

```python
///Grouping by the amount of purchases 
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
From the above two graphs we conclude that more number of females are buying and purchasing power of females is greater 
```

```python
///our next analysis will be on Age group 
```

```python
ax=sns.countplot(x='Age Group',hue='Gender',data=df)
for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum()
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
Conclusion : Out major Customers are Females from the age 26-35 , followed by 36-45 
```

```python
////Checking total number of orders state wise 
```

```python
state_orders=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).head(10)
sns.barplot(x='State',y='Orders',data=state_orders)
sns.set(rc={'figure.figsize':(15,50)})
```

```python
checking amount wise order's rathar than quantity , You will see that kerala was in 8th position but amount wise it falls from top 10 
```

```python
state_orders=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(10)
sns.barplot(x='State',y='Amount',data=state_orders)
sns.set(rc={'figure.figsize':(5,10)})
```

```python
//Conclusion - Most of the orders are from Uttar Pradesh , Maharashathra, karnataka and their purchasing power is greater as well 
```
